```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Dataset importing
df = pd.read_csv("D:\Imarticus\PROJECTS\SOCIAL MEDIA ANALYTICS\DATASETS\Social Media Analytics 2024 Noisy un clear dataset.csv")
# General things to do 
df.info()
df.describe(include='all')
```

```python
# Cleaning the column names first
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
df.columns
```

```python
# Checking duplicates
df.duplicated().sum()
# Drop duplicates
df = df.drop_duplicates()
df
```

```python
# CONVERTING THE DATA TYPES OF THE DATASET 
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns
df
```

```python
#FILLING THE NA VALUES 
for col in num_cols:
    if df[col].isna().sum() > 0:
        df[col] = df[col].fillna(df[col].median())
df
```

```python
# FILLING THE CATEGORICAL VALUES USING MODE
for col in cat_cols:
    if df[col].isna().sum() > 0:
        df[col] = df[col].fillna(df[col].mode()[0])
df
```

```python
# Convert date columns
df['post_date'] = pd.to_datetime(df['post_date'], errors='coerce')
# If conversion created NaT, fill again
df['post_date'] = df['post_date'].fillna(df['post_date'].mode()[0])
df
```

```python
# Checking the outliers
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    df[col] = np.where(df[col] < lower, lower, df[col])
    df[col] = np.where(df[col] > upper, upper, df[col])
df
```

```python
# Example checks
df = df[df['likes'] >= 0]
df = df[df['shares'] >= 0]
df = df[df['comments'] >= 0]
df
```
